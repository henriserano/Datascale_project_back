```python
import random
from faker import Faker
import pandas as pd

# Initialiser Faker
fake = Faker()

# Définir les rôles
roles = ["Employé", "Manager", "Administrateur"]

# Définir les départements métier et leurs descriptions
departments = {
    "Human Resources": "Employee management, recruitment, and workplace policies",
    "Information Technology": "Technical infrastructure, software development, and digital solutions",
    "Finance & Accounting": "Financial planning, accounting, and budget management",
    "Supply Chain Management": "Procurement, logistics, and inventory management",
    "Quality Control": "Quality assurance and product testing",
    "Customer Service": "Customer support and satisfaction management",
    "Sales": "Sales operations and client relationship management",
    "Marketing": "Brand management and marketing campaigns",
    "Research & Development": "Product innovation and development",
    "Product Management": "Product lifecycle and roadmap management",
    "Legal": "Legal compliance and contract management",
    "Facilities Management": "Workplace infrastructure and maintenance",
    "Manufacturing": "Production and assembly operations",
    "Engineering": "Technical design and implementation"
}

def generate_email(name, professional=True):
    """Génère une adresse email professionnelle ou personnelle."""
    domain = "process-boost.com" if professional else fake.free_email_domain()
    email = f"{name.lower().replace(' ', '.')}@{domain}"
    return email

def generate_fake_user():
    """Génère un utilisateur fictif avec toutes les informations nécessaires."""
    first_name = fake.first_name()
    last_name = fake.last_name()
    full_name = f"{first_name} {last_name}"
    professional_email = random.choice([True, False])
    email = generate_email(full_name, professional=professional_email)
    role = random.choice(roles)
    department = random.choice(list(departments.keys()))
    
    return {
        "First Name": first_name,
        "Last Name": last_name,
        "Email": email,
        "Role": role,
        "Department": department
    }

def generate_fake_users(n):
    """Génère une liste de n utilisateurs fictifs."""
    users = [generate_fake_user() for _ in range(n)]
    return pd.DataFrame(users)

# Générer et afficher 10 utilisateurs fictifs
fake_users = generate_fake_users(100)
print(fake_users)
```

```python
import boto3
import random
from faker import Faker
from datetime import datetime, timedelta

# AWS Credentials
AWS_SECRET = "ukfCjJXm2VmVu8C8jH1kgK3gST+haWU3JgCWrbR2"
AWS_ID = "AKIAT6SNFGTUYMXEPNUQ"
REGION = "us-east-1"

# Initialize Faker
fake = Faker()

# Initialize DynamoDB resource
dynamodb = boto3.resource(
    'dynamodb',
    aws_access_key_id=AWS_ID,
    aws_secret_access_key=AWS_SECRET,
    region_name=REGION
)

# Tables
users_table = dynamodb.Table('Datascale_users')
requests_table = dynamodb.Table('Requests')

# Sample request statuses and descriptions
statuses = ["Pending", "Approved", "Rejected"]
sectors = [
    "Human Resources", "Information Technology", "Finance & Accounting",
    "Supply Chain Management", "Quality Control", "Customer Service",
    "Sales", "Marketing", "Research & Development", "Product Management",
    "Legal", "Facilities Management", "Manufacturing", "Engineering"
]

# Generate fake requests
def generate_fake_requests(n=100):
    # Fetch all users from the Users table
    users_response = users_table.scan()
    users = users_response.get('Items', [])
    
    if not users:
        raise ValueError("No users found in the Datascale_users table.")
    
    requests = []
    for _ in range(n):
        user = random.choice(users)
        sector = random.choice(sectors)
        status = random.choice(statuses)
        submitted_on = fake.date_between(start_date="-1y", end_date="today")
        description = f"Request for {sector.lower()} services for {user['First Name']} {user['Last Name']}."
        
        # Generate additional details dynamically
        additional_details = {
            "Position": fake.job(),
            "Start Date": fake.date_between(start_date="+1d", end_date="+30d").strftime("%Y-%m-%d")
        }
        
        request = {
            "RequestID": str(fake.uuid4()),
            "UserID": user["ID_Users"],
            "Sector": sector,
            "Status": status,
            "SubmittedOn": submitted_on.strftime("%Y-%m-%d"),
            "Description": description,
            "AdditionalDetails": additional_details
        }
        
        requests.append(request)
    
    return requests

# Push requests to DynamoDB
def push_requests_to_dynamodb(requests):
    for request in requests:
        requests_table.put_item(Item=request)

# Main function to generate and push requests
def main():
    try:
        print("Generating fake requests...")
        fake_requests = generate_fake_requests(n=100)
        
        print("Pushing requests to DynamoDB...")
        push_requests_to_dynamodb(fake_requests)
        
        print("Fake requests successfully added to the Requests table.")
    except Exception as e:
        print(f"An error occurred: {e}")

# Execute the script
if __name__ == "__main__":
    main()
```
